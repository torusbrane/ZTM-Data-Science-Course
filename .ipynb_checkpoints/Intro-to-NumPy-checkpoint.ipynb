### Standard Deviation and Variance
#### the measure of the spread of data

```python
# standard deviataion = a measure of how spread out a group of numbers is from the mean
np.std(a2)
```

```python
# variance = the measure of the avg degree in which each number is different to the mean
# higher variance = wider range of numbers
# lower variance = lower rand of numbers
np.var(a2)
```

```python
# the square root of the variance is the standard deviation
np.sqrt(np.var(a2))
```

```python
np.std(a2)
```

```python
mat1
```

```python
mat1.shape, mat2.shape
```

```python
#Element-wise mulitplication
mat1 * mat2
```

```python
#Dot Product
#np.dot(mat1, mat2) - this won't work because the outside numbers don't make up the new
#size of the answer, in this case, (5, 3)
```

# Dot Product Example (Nut Butter Sales)

```python
np.random.seed(0)
#number of jars sold
sales_amount = np.random.randint(20, size=(5,3))
sales_amount
```

```python
weekly_sales = pd.DataFrame(sales_amount, index=
                            ["Mon", "Tues", "Wed", "Thurs", "Fri"],
                            columns=["Almond Butter", "Peanut Butter", "Cashew Butter"])
weekly_sales
```

```python
prices = np.array((10, 8, 12))
prices
```

```python
prices.shape
```

```python
#create butter prices dataframe
butter_prices = pd.DataFrame(prices.reshape(1,3), index=["Price"],
                             columns=["Almond Butter", "Peanut Butter", "Cashew Butter"])
butter_prices                             
```

```python
#shapes aren't aligned so we need to transpose
total_sales = prices.dot(sales_amount.T)
total_sales
```

```python
#create daily_sales
butter_prices.shape, weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
